```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('/kaggle/input/diwali-sales/Diwali Sales Data.csv',encoding='unicode_escape')
df
```

```python
df.info()
```

drop columns having null values

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.drop_duplicates()
```

```python
df.info()
```

```python
ax=sns.countplot(df,x='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
bx=sns.barplot(sales_gen,x='Gender',y='Amount')
bx.bar_label(bx.containers[0])
```

# from the above we can see that most of the buyers are females and purchasing power of female is also grater than male

Age

```python
cx=sns.countplot(df,x='Age Group',hue='Gender')
for bars in cx.containers:
    cx.bar_label(bars)
```

```python
#total amount vs age group
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(sales_age,x='Age Group',y='Amount')
```

From above graph we can see that most of the buyers are from 26-35 age group

# State

```python
df['State'].unique()
```

```python
#total number of orders from top 10 states
sales_state=df.groupby(by='State',as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(sales_state,x='State',y='Orders')
```

```python
#total amount of sales from top 10 states
amount_states=df.groupby(by='State',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(amount_states,x='State',y='Amount')
```

From the above we can see that most of the orders are from Uttar pradesh Mahrashtra and Karnatka respectively 
most amount of sales also came from same States

# Marrital status

```python
dx=sns.countplot(df,x='Marital_Status')
sns.set(rc={'figure.figsize':(9,5)})
for bars in dx.containers:
    dx.bar_label(bars)
```

```python
sales_state=df.groupby(by=['Gender','Marital_Status'],as_index=False)['Amount'].sum().sort_values(by=['Amount'],ascending=False)
sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(sales_state,x='Marital_Status',y='Amount',hue='Gender')
```

From above we can see that most of the purchase done by married woemen and they have higher purchasing power

# Occupation

```python
dx=sns.countplot(df,x='Occupation')
sns.set(rc={'figure.figsize':(2,5)})
for bars in dx.containers:
    dx.bar_label(bars)
```

```python
sales_occupation=df.groupby(by='Occupation',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(sales_occupation,x='Occupation',y='Amount')
```

From above we can see that It sector Healthcare and aviation occupations have high purcasing power

# Product Category

```python
fx=sns.countplot(df,x='Product_Category')
sns.set(rc={'figure.figsize':(20,5)})
for bars in fx.containers:
    fx.bar_label(bars)
```

```python
sales_occupation=df.groupby(by='Product_Category',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(sales_occupation,x='Product_Category',y='Amount')
```

from above we can see that most sold products are food,Clothing and apparel and electronic and gadgets
